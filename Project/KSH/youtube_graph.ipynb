```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import textwrap

# sns 스타일 설정
sns.set_style("white")

# 라벨 자동 줄바꿈 함수
def wrap_labels(labels, width=5):
    return [textwrap.fill(label, width) for label in labels]

# 선 그래프 그리기
plt.figure(figsize=(20, 6))
lineplot = sns.lineplot(data=game_data, x='게임 이름', y='조회수', color="darkred")

# 데이터 포인트에 라벨 추가
for x, y in zip(game_data['게임 이름'], game_data['조회수']):
    plt.text(x, y, str(y), color='lightcoral', ha='right')

# 그래프 제목 설정 및 글자 크기 조절
plt.title('', fontsize=16)

# x축 라벨 회전 및 글자 크기 조절, 자동 줄바꿈 적용
wrapped_labels = wrap_labels(game_data['게임 이름'], width=10)
plt.xticks(ticks=range(len(game_data['게임 이름'])), labels=wrapped_labels, rotation=0, fontsize=7)

# 상단 및 우측 테두리 제거
sns.despine()

# 그래프 보여주기
plt.show()
```
